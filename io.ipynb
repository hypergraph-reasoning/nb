```python
# InstalaÃ§Ã£o das bibliotecas necessÃ¡rias
!pip install networkx matplotlib numpy

# ImportaÃ§Ã£o das bibliotecas
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

# CriaÃ§Ã£o de um hipergraph bÃ¡sico utilizando NetworkX
G = nx.Graph()

# Adicionando nÃ³s ao hipergraph
G.add_node("A", label="Node A")
G.add_node("B", label="Node B")
G.add_node("C", label="Node C")

# Adicionando hiperarestas (representadas por arestas que conectam mÃºltiplos nÃ³s)
# Hiperaresta AB: conecta A e B
G.add_edge("A", "B", label="Hyperedge AB")

# Hiperaresta ABC: conecta A, B e C
G.add_edge("A", "C", label="Hyperedge ABC")
G.add_edge("B", "C", label="Hyperedge ABC")

# Exibir os nÃ³s e arestas do hipergraph
print("NÃ³s do Hipergraph:", G.nodes(data=True))
print("Arestas do Hipergraph:", G.edges(data=True))

# Definir uma funÃ§Ã£o para desenhar o hipergraph
def draw_hypergraph(G):
    pos = nx.spring_layout(G)  # PosiÃ§Ã£o dos nÃ³s no layout de mola

    # Desenhar os nÃ³s
    nx.draw_networkx_nodes(G, pos, node_size=700, node_color="lightblue")

    # Desenhar as arestas
    nx.draw_networkx_edges(G, pos, edgelist=G.edges(), width=2, edge_color="gray")

    # Adicionar rÃ³tulos aos nÃ³s e arestas
    nx.draw_networkx_labels(G, pos, font_size=12, font_family="sans-serif")
    edge_labels = nx.get_edge_attributes(G, 'label')
    nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

    # Mostrar a visualizaÃ§Ã£o
    plt.title("VisualizaÃ§Ã£o do Hipergraph")
    plt.axis("off")  # Desativar os eixos
    plt.show()

# Chamar a funÃ§Ã£o para desenhar o hipergraph
draw_hypergraph(G)
```
